```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
import tensorflow as tf
import pyswarms as ps
import seaborn as sns

# Load the data
df = pd.read_csv(r'D:\Y\DATA\PREPROCESS\BookA.csv', parse_dates=['Date'], index_col=['Date'])

# Define the AQI calculation functions

def calcPM25(x):
    if x <= 10:
        return 1
    elif x >=11 and x <= 20:
        return 2
    elif x >=21 and x <= 25:
        return 3
    elif x >=26 and x <= 50:
        return 4
    elif x >=51 and x <= 75:
        return 5
    elif x >=76:
        return 6
    else:
        return 1
    
    
## CO Sub-Index calculation
def calcCO(x):
    if x <= 1:
        return x * 50 / 1
    elif x <= 2:
        return 50 + (x - 1) * 50 / 1
    elif x <= 10:
        return 100 + (x - 2) * 100 / 8
    elif x <= 17:
        return 200 + (x - 10) * 100 / 7
    elif x <= 34:
        return 300 + (x - 17) * 100 / 17
    elif x > 34:
        return 400 + (x - 34) * 100 / 17
    else:
        return 0
    
# Apply AQI calculation functions to create additional AQI columns
df["co_AQI"] = df["CO"].apply(lambda x: calcCO(x))

df["pm25_AQI"] = df["PM2.5"].apply(lambda x: calcPM25(x))
# Apply other AQI calculations for remaining features

# Convert the dataframe to float32
df = df.astype(np.float32)

# Visualize the correlation matrix
plt.figure(figsize=(15, 15))
cor = df.corr(method='pearson')
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()

# Split the data into features and targets
features = df.values
targets = df[["PM2.5"]].values

# Split the data into train, validation, and test sets
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=1, shuffle=False)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1, shuffle=False)

# Scale the data using MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)

# Define the window length, batch size, and number of features
win_length = 11
batch_size = 4
num_features = features.shape[1]

# Create TimeseriesGenerators for train, validation, and test data
train_generator = TimeseriesGenerator(X_train, y_train, length=win_length, sampling_rate=1, batch_size=batch_size)
val_generator = TimeseriesGenerator(X_val, y_val, length=win_length, sampling_rate=1, batch_size=batch_size)
test_generator = TimeseriesGenerator(X_test, y_test, length=win_length, sampling_rate=1, batch_size=batch_size)

# Define the CNN-LSTM model
model_CNNLSTM = tf.keras.Sequential([
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True), input_shape=(win_length, num_features)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Conv1D(filters=128, kernel_size=2, activation=tf.keras.layers.LeakyReLU(alpha=0.5)),
    tf.keras.layers.MaxPooling1D(pool_size=1),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(targets.shape[1])
])

# Compile the model
model_CNNLSTM.compile(optimizer=tf.keras.optimizers.Adadelta(learning_rate=0.001), loss=tf.losses.MeanSquaredLogarithmicError())

# Define the PSO optimization function
def pso_optimization(model, train_generator, val_generator, num_particles, max_iterations):
    def fitness_function(params):
        weights = np.split(params, [64, 256, 320, 416])

        model.layers[0].set_weights([weights[0]])

        model.layers[1].layers[0].set_weights([weights[1], weights[2], weights[3]])
        model.layers[1].layers[1].set_weights([weights[1], weights[2], weights[3]])

        model.layers[2].set_weights([weights[4]])

        history = model.fit(train_generator, validation_data=val_generator, epochs=1, verbose=0)
        val_loss = history.history['val_loss'][0]
        return val_loss

    num_weights = sum([w.size for w in model.get_weights()])

    optimizer = ps.single.GlobalBestPSO(n_particles=num_particles, dimensions=num_weights, options={'c1': 0.5, 'c2': 0.3, 'w': 0.9})

    best_cost, best_pos = optimizer.optimize(fitness_function, iters=max_iterations)

    return best_pos

# Set the number of particles and maximum iterations for PSO
num_particles = 11
max_iterations = 50

# Perform PSO optimization to find the best weights
optimized_weights = pso_optimization(model_CNNLSTM, train_generator, val_generator, num_particles, max_iterations)

# Set the best weights obtained from PSO to the model
weights = np.split(optimized_weights, [64, 256, 320, 416])
model_CNNLSTM.layers[0].set_weights([weights[0]])
model_CNNLSTM.layers[1].layers[0].set_weights([weights[1], weights[2], weights[3]])
model_CNNLSTM.layers[1].layers[1].set_weights([weights[1], weights[2], weights[3]])
model_CNNLSTM.layers[2].set_weights([weights[4]])

# Evaluate the model on the test data
test_loss = model_CNNLSTM.evaluate(test_generator)
print("Test Loss:", test_loss)

# Make predictions using the model
predictions = model_CNNLSTM.predict(test_generator)

# Invert the scaling of the predictions and actual values
predictions = scaler.inverse_transform(predictions)
actual_values = scaler.inverse_transform(y_test[win_length:])

# Plot the predicted and actual values
plt.plot(actual_values, label="Actual")
plt.plot(predictions, label="Predicted")
plt.legend()
plt.show()
```
