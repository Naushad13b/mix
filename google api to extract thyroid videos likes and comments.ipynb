```python
import googleapiclient.discovery
import csv

# Set your YouTube Data API key
API_KEY = 'AIzaSyAdbh_xyl_Sb-RcFrZhPnXhVUMgEPG7mwI'

# Initialize the YouTube Data API client
youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=API_KEY)

# Search for videos related to "thyroid"
search_response = youtube.search().list(
    q='thyroid',  # Your search query here
    type='video',  # You can also search for channels or playlists
    part='id,snippet',  # Include 'snippet' to get video titles
    maxResults=100,  # Maximum number of results per request (you can change this)
).execute()

# Get the total number of videos related to "thyroid"
total_videos_related_to_thyroid = search_response['pageInfo']['totalResults']

# Initialize counters for total likes, dislikes, and comments
total_likes = 0
total_dislikes = 0
total_comments = 0

# Create a CSV file to store the collected data
csv_file = open('thyroid_videos_statistics.csv', mode='w', encoding='utf-8', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Video ID', 'Title', 'Likes', 'Dislikes', 'Comments'])

# Iterate through the search results and retrieve statistics for each video
for item in search_response['items']:
    video_id = item['id']['videoId']
    video_title = item['snippet']['title']
    
    # Call the API to retrieve video statistics
    response = youtube.videos().list(
        part='statistics',
        id=video_id
    ).execute()
    
    # Extract statistics from the response
    video_statistics = response['items'][0]['statistics']
    likes = int(video_statistics.get('likeCount', 0))
    dislikes = int(video_statistics.get('dislikeCount', 0))
    comments = int(video_statistics.get('commentCount', 0))
    
    # Add to total counters
    total_likes += likes
    total_dislikes += dislikes
    total_comments += comments
    
    # Write statistics to the CSV file
    csv_writer.writerow([video_id, video_title, likes, dislikes, comments])

    # Print statistics for the current video
    print(f"Video ID: {video_id}")
    print(f"Title: {video_title}")
    print(f"Likes: {likes}")
    print(f"Dislikes: {dislikes}")
    print(f"Comments: {comments}")
    print("-" * 20)

# Close the CSV file
csv_file.close()

# Print the total statistics for all videos
print(f"Total videos related to thyroid: {total_videos_related_to_thyroid}")
print(f"Total Likes: {total_likes}")
print(f"Total Dislikes: {total_dislikes}")
print(f"Total Comments: {total_comments}")
```
